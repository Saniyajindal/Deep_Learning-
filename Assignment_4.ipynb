```python
import pandas as pd
import numpy as np

# --- STEP 1: Dataset Creation (Kaggle dataset structure ke mutabiq) ---
# Chunki aapke paas file nahi hai, hum wahi data manually create kar rahe hain [cite: 11, 14]
raw_data = {
    'Years Experience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7],
    'TestScore': [90, 80, 75, 95, 88, 85, 92, 84, 98, 91],
    'Salary': [39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189]
}

# CSV file save karna aur load karna [cite: 31, 44]
df = pd.DataFrame(raw_data)
df.to_csv("multiple_linear_regression_dataset.csv", index=False)
data = pd.read_csv("multiple_linear_regression_dataset.csv")

print("Dataset Loaded Successfully:")
print(data.head()) # Pehli 5 rows [cite: 33, 46]

# --- STEP 2: Separate Inputs and Output ---
# X mein input features aur y mein target variable [cite: 57, 58, 65, 67]
X = data[["Years Experience", "TestScore"]].values
y = data["Salary"].values

# --- STEP 3: Initialize Model Parameters ---
# Weights aur bias ko zero se shuru karna [cite: 78, 88, 89]
n_features = X.shape[1]
w = np.zeros(n_features)
b = 0.0

# --- STEP 4-7: Define Functions for Training ---
def predict(X, w, b):
    # Linear model: y_hat = w1x1 + w2x2 + b [cite: 99, 108]
    return X.dot(w) + b

def mean_squared_error(y, y_hat):
    # Mean Squared Error formula implementation [cite: 119, 131]
    return ((y_hat - y) ** 2).mean()

def compute_gradients(X, y, y_hat):
    # MSE ke gradients w.r.t weights aur bias [cite: 155, 157]
    N = len(y)
    dw = (2/N) * X.T.dot(y_hat - y)
    db = (2/N) * (y_hat - y).sum()
    return dw, db

def update_parameters(w, b, dw, db, lr):
    # Gradient Descent update rule [cite: 175, 176]
    w = w - lr * dw
    b = b - lr * db
    return w, b

# --- STEP 8: Training Loop ---
# Learning rate aur epochs set karke train karna [cite: 189, 192]
lr = 0.0001
epochs = 2000

print("\nStarting Training...")
for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.2f}")

# --- STEP 9: Final Evaluation ---
print("\nFinal Parameters:")
print(f"Weights: {w}")
print(f"Bias: {b:.4f}")

# Naye data par prediction [cite: 211, 212]
new_candidate = np.array([4.5, 68])
prediction = new_candidate.dot(w) + b
print(f"\nPredicted Salary for (Exp: 4.5, Score: 68): {prediction:.2f}")
```
