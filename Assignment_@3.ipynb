```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- 1. Dataset Generation (Synthesizing Glass Data) ---
# Since glass.csv is missing, we create synthetic numeric measurements
np.random.seed(42)
data_size = 214
# Simulating 9 glass features (RI, Na, Mg, Al, Si, K, Ca, Ba, Fe)
X_synthetic = np.random.randn(data_size, 9)
y_synthetic = np.random.randint(0, 2, data_size)

# --- 2. Data Preparation ---
# Splitting into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(X_synthetic, y_synthetic, test_size=0.2)

# Scaling is crucial because features have different ranges
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# --- 3. Core Functions (TODOS from Manual) ---

def sigmoid(z):
    # Converts score z into probability p between 0 and 1
    return 1 / (1 + np.exp(-z))

def predict_proba(X, w, b):
    # Computes evidence score z and converts to probability p
    z = X @ w + b
    return sigmoid(z)

def loss(y, p):
    # Binary Cross Entropy penalizes wrong confidence
    return -np.mean(y * np.log(p + 1e-15) + (1 - y) * np.log(1 - p + 1e-15))

def update_weights(X, y, w, b, lr):
    p = predict_proba(X, w, b)
    error = p - y # Correction signal
    w = w - lr * (X.T @ error) / len(y)
    b = b - lr * np.mean(error)
    return w, b

# --- 4. Training Loop ---
w = np.zeros(X_train.shape[1])
b = 0.0
lr = 0.1
epochs = 100

for i in range(epochs):
    w, b = update_weights(X_train, y_train, w, b, lr)

# --- 5. Evaluation ---
p_test = predict_proba(X_test, w, b)
# Thresholding: 0.5 is balanced, 0.7 is more conservative
acc_05 = np.mean((p_test >= 0.5) == y_test)
acc_07 = np.mean((p_test >= 0.7) == y_test)

print(f"Final Test Loss: {loss(y_test, p_test):.4f}")
print(f"Accuracy at 0.5 threshold: {acc_05*100:.2f}%")
print(f"Accuracy at 0.7 threshold: {acc_07*100:.2f}%")
```
